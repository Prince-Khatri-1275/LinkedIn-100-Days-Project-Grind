```python
import os
import cv2
import numpy as np
import mediapipe as mp

drawing_utils = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
PATH_TO_VIDEO = os.path.join("pushup_vids", "pushups.mp4")
```

```python
!pip install opencv-jupyter-ui
from opencv_jupyter_ui import *
```

```python
vid = cv2.VideoCapture(PATH_TO_VIDEO)
scaling_factor = 0.25
while True:
    ret, frame = vid.read()

    fh, fw, _ = frame.shape # as it is most clost to an Numpy Array
    frame = cv2.resize(frame, (int(fw*scaling_factor), int(fh*scaling_factor)))

    cv2_imshow("Frame", frame)
    
    key = cv2_waitKey(1)
    if key == ord("q"):
        break

vid.release()
cv2_destroyAllWindows()
```
